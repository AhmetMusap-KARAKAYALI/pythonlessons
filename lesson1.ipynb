Karakter dizilerinin metodları

```python
string='ahmet musap'
```

```python
print(string[1])
#harflerin numaralandırılması sıfırdan başlar
```

```python
print(string[0:4])
#sıfırdan dördüncü harfe kadar olan tüm harfleri kapsar dördüncü harf hariç
```

```python
print(string[0:len(string):2])
""" burada ise yaptığımız şey string[start:stop:step] yani start başlangıç değeri stop bitiş değeri step ise yazılan hharf kadar atlayarak yazar
 bu kodda iki yazdığımızda her iki harften birini alıyor"""
```

```python
print(string.lower())
print(string.upper())
"""lower tüm harfleri küçük yapar
upper da tam tersi tüm harfleri büyyük yapar"""
```

```python
print(string)

string = string.replace('musap', 'ibrahim')
print(string)

string = string.replace('ahmet', 'mehmet')
print(string)
#replace yazdığımız zaman ilk kelimenin yerine ikinci yazdığımız kelimeyi koyar
```

```python
print("A".islower())
print("A".isupper())
```
